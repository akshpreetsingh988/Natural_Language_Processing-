## Q4: Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men , women , and people in each document. What has happened to the usage of these words over time?

```python
#Downloading the state_union corpus
nltk.download('state_union')
```

```python
from nltk.corpus import state_union
```

```python
print(state_union.fileids())
```

```python
%matplotlib inline
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in state_union.fileids()
    for w in state_union.words(fileid)
    for target in ['men', 'women','people']
    if w.lower().startswith(target))
```

```python
cfd
```

```python
cfd.plot()
```

## Q7: Write a program to generate a table of lexical diversity scores (i.e., token/type ratios) for each genre. Include the full set of Brown Corpus genres ( nltk.corpus.brown.categories() ). Which genre has the lowest diversity

```python
brown_genres = brown.categories()
brown_genres
```

```python
print("Genre , Number of tokens , NUmber of Types , Lexical Diversity in the Genre")
for _ in brown_genres :
    categories = brown.words(categories = _)
    print( _ , len(categories) , len(set(categories)) , len(categories)/len(set(categories))*1.0)
```

According to the output we see science fiction has the lowest lexical diversity

## Q8: Write a function that finds the 50 most frequently occurring words of a text?

```python
freq_dist = nltk.FreqDist(brown.words(categories = 'news'))
num=1
for i in freq_dist.most_common(50):
    print(num , i)
    num=num+1
```
