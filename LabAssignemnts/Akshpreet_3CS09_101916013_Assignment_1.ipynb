## Q1: Find the collocations in text5

```python
text5
```

```python
# Collocations 
# colocations are words occuring frequently together ex look after
#found using different metrics like t test, chi squared , point wise mutual informatio n

text5.collocations()
```

```python
# Collocations are expressions of multiple words which commonly co-occur.
from nltk.collocations import*
```

```python
bigram_measure = nltk.collocations.BigramAssocMeasures() #Bigram
finder = BigramCollocationFinder.from_words(text5)
finder.nbest(bigram_measure.pmi , 100)
```

```python
trigram_measure = nltk.collocations.TrigramAssocMeasures() #Trigram
finder = TrigramCollocationFinder.from_words(text5)
finder.nbest(trigram_measure.pmi , 100)
```

```python
fourgram_measures = nltk.collocations.QuadgramAssocMeasures()
finder = QuadgramCollocationFinder.from_words(text5)
finder.nbest(fourgram_measures.pmi , 100)
```
