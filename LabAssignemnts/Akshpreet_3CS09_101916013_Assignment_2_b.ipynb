### Q3 .Define a conditional frequency distribution over the Names corpus that allows you to see which initial letters are more frequent for males vs. females

```python
Conditional_freq = nltk.ConditionalFreqDist(
                    (fileid , name[0])
                    for fileid in names.fileids()
                    for name in names.words(fileid)
                    )
```

```python
Conditional_freq
```

```python
Conditional_freq.plot()
```

For females , the letter 'M' has the highest frequency of initial letter and letter 'X' has the lowest .
For Males ,  the letter 'S' has the highest frequency of initial letter and letter 'X' has the lowest .

### Q4 .Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.


```python
def supergloss(s):
    definitions=s.definition();
    for hypo in s.hyponyms():
        definitions+="\n"+hypo.definition()
    for hyper in s.hypernyms():
        definitions+="\n"+hyper.definition()
    return definitions
```

```python
definitions = supergloss(wn.synset('car.n.01'))
print(definitions)
```

### Q5 . The polysemy of a word is the number of senses it has. Using WordNet, we can determine that the noun dog has 7 senses with: len(wn.synsets('dog', 'n')). Compute the average polysemy of nouns, verbs, adjectives and adverbs according to WordNet.


```python
from __future__ import division
```

```python
from nltk.corpus import wordnet as wn
```

```python
from collections import defaultdict
```

```python
word_by_pos = defaultdict(set)
```

```python
for synset in wn.all_synsets():
    pos = synset.pos()
    for lemma in synset.lemmas():
        word_by_pos[pos].add(lemma)
```

```python
import nltk
from nltk.corpus import wordnet as wn
```

```python
nouns = list(wn.all_synsets('n'))
adjectives = list(wn.all_synsets('a'))
verbs = list(wn.all_synsets('v'))
adverbs = list(wn.all_synsets('r'))
```

```python
def words(synset) :
    all_words = []
    for syn in synset : 
        all_words += syn.lemma_names()
    total = len(set(all_words))
    return total
```

```python
def polysemy(synset):
    meaning = sum( 1 for x in synset )
    return meaning 
```

```python
def average_polysemy(synset):
    average = polysemy(synset) / words(synset)
    return average
```

```python
print("Average polysemy of Nouns " , average_polysemy(nouns))
```

```python
print("Average polysemy of Adjectives " , average_polysemy(adjectives))
```

```python
print("Average polysemy of Verbs " , average_polysemy(verbs))
```

```python
print("Average polysemy of Adverbs " , average_polysemy(adverbs))
```
